```python
# 22

mean_age_0 = ds[ds['target'] == 0]['age'].mean()
mean_age_1 = ds[ds['target'] == 1]['age'].mean()
print(f"mean age in ill persons : {mean_age_1:.2f}")
print(f"mean age in healthy persons: {mean_age_0:.2f}")

plt.figure(figsize=(10, 6))
plt.hist(ds[ds['target'] == 0]['age'], bins=15, alpha=0.5 ,  label=" not ill")
plt.hist(ds[ds['target'] == 1]['age'], bins=15, alpha = 0.5 , label='ill')
plt.xlabel('age')
plt.ylabel('frequency')
plt.title('Age distribution based on frequency')
plt.legend()
plt.show()
# The average age in the two groups is very close and there is no significant difference.
# Therefore, in this dataset, age alone is not a determining factor for disease. Although in the general population the risk increases with age, this pattern may not be seen in this sample due to data size limitations.
```
